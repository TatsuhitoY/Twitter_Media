```python
#セクション1

#Twitter解析のコード
import tweepy
# Twitter認証コードは別ファイルで管理

consumer_key ='Confidential'
consumer_secret ='Confidential'
access_token='Confidential'
access_secret ='Confidential'

# Twitterオブジェクトの生成
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth, wait_on_rate_limit = True)

import datetime

# JSTとUTCの差分
DIFF_JST_FROM_UTC = 9
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
import csv
import numpy as np

def job():
  url = "https://pressgallery.house.gov/member-data/members-official-twitter-handles"
  rest = requests.get(url)
  soup = BeautifulSoup(rest.text, "html.parser")
  data_list = soup.find_all(class_ = "field-item even")
  print(type(data_list))
  print(data_list)

  article_list = [data.text for data in data_list]
  # for i in data_list:
  #   if i.attrs["class_"] == "x165":
  #   data_list.remove(i)
  print(article_list)

job()
```
